```python

# Count the occurrences of 0 and 1 in the 'churn' column
churn_counts = client_df['churn'].value_counts()

# Extract the counts for 0 and 1
churn_0_count = churn_counts[0]
churn_1_count = churn_counts[1]

# Create a bar chart
plt.bar([0, 1], [churn_0_count, churn_1_count], tick_label=['False', 'True'])

# Add labels and title
plt.xlabel('Churn Status')
plt.ylabel('Count')
plt.title('Churn Status Count')

print("Customer Churn Count:",churn_1_count)
print("Customer Retain Count:",churn_0_count)
# Show the plot
plt.show()
```

```python
import numpy as np
# Count the occurrences of 0 and 1 in the 'churn' column
churn_counts = client_df['churn'].value_counts()

# Extract the percentage for 0 and 1
churn_0_per = (churn_counts[0]/churn_counts.sum())*100
churn_1_per = (churn_counts[1]/churn_counts.sum())*100 

pielabels=['retained','churned']

# Create a bar chart
plt.pie(churn_counts,labels=pielabels)

plt.title('Churn Status Percentage')

print("Customer Churn Percentage:",churn_1_per)
print("Customer Retain Percentage:",churn_0_per)
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Filter records where churn value is 1
churn_true_client_df = client_df[client_df['churn'] == 1]

# Plot boxplot for cons_gas_12m column
plt.boxplot(churn_true_client_df['cons_gas_12m'])

# Set labels and title
plt.xlabel('Churn = 1')
plt.ylabel('cons_gas_12m')
plt.title('Boxplot of cons_gas_12m for Churn = 1')

# Format y-axis in 1e6 format
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.0f}e6'.format(x * 1e-6)))

# Show the plot
plt.show()
```
